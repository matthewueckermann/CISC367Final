## Background

Various industries release or cause the release of chemicals and particulates which have the potential to directly and severely impact the health and lives of those in the surrounding community. Releases of toxic chemicals are often mitigated to some extent to meet local air pollution licensing requirements and to prevent negative press attention. However, due to the nature of the equipment used as well as the effectiveness of mitigation technology, some releases will occur. While other industrial activities, like [mountain removal mining](https://en.wikipedia.org/wiki/Mountaintop_removal_mining#Health_impacts), cause the release of large amounts of particulates as well as toxic chemicals which cannot be easily controlled.

The fact that industries pollute chemicals which have been shown to be toxic in lab settings or which are believed to cause lung disorders, begs the question of whether these emissions have demonstrable effects on markers of community health. Do elevated quantities of exposure to these chemicals cause higher cancer rates? Or are chemicals emitted in small enough quantities that other factors dominate whether or not people develop certain disorders. This information equips communities with the knowledge about whether they should be concerned about surrounding industrial activity as well the extent to push for greater emission control or use of safer practices.

## Research Questions

- To what extent can the cities with higher than average rates of cancer be predicted by looking at carcinogen release from facilities in the same county?
- Do any routes of chemical release (air, water, landfill) cause a greater associated rate of cancer?
- To what extent can cities with higher than average rates of asthma be predicted by looking at rates of nitrogen oxide and particulate matter emissions?
- Do communities with a more active mining sector have a greater associated rate of asthma then those that do not?

## Justification

For the first question, carcinogens are by [definition](https://www.cancer.gov/about-cancer/causes-prevention/risk/substances/carcinogens) compounds which are believed to cause cancer. Exposure limits have historically been defined in terms of [occupational exposure limits](https://www.cdc.gov/niosh/npg/nengapdxa.html), as these are where people are generally exposed to large enough quantities of these compounds frequently enough for them to make a difference. However, it is less straightforward to define exposure quantities which could cause surrounding communities to develop cancer at higher rates. Some studies show that atmospheric carcinogens in the UK have been linked with higher child cancer rates [1]. Completing a similar study in the US could give residents important knowledge about the facilities that surround them. 

The second question then builds on this to see if there is a noticeable difference based on emission route. As my midterm showed that the majority of emissions are emitted to the air, I was interested if the amount emitted to the water, impacts health in a more detrimental way.

For the third question, it is widely accepted that exposure to particulates as well as oxides can trigger asthma attacks and may also lead to the development of new-onset asthma in some people [2]. It would be interesting to characterize whether emissions from industrial facilities play a role in exacerbating this condition, or if the particulates and other compounds which trigger asthma can be primarily attributed to transportation - another major source of these compounds. This could equip residents with extra knowledge of what to attribute their asthma too and what types of facilities to avoid to prevent triggering it.

The fourth question builds off of the third, by investigating the impact of mining specifically on asthma. This is because some forms of mining, namely mountain strip top, have been associated with increased amount of respiratory disorders, including asthma [3]. This builds on the history of lung disorders associated with mining, notably [black lung](https://en.wikipedia.org/wiki/Coalworker%27s_pneumoconiosis) in coal miners. This could show that entire communities, not just workers, are impacted by exposure brought about by the mining industry.

## Ethical Concerns and Considerations

    - Notably also, this analysis assumes that all residents in a city are exposed to emissions in their city at the same level, which is most definitely not the case. As one moves further away from the emissions point source, the concentration will rapidly decline. As such, people who live closer to a site (who are most likely come from less privileged communities) are at much more risk than people on the other side of the city. Because of this the major stakeholders mentioned above cannot jump to any conclusions, specifically the residents stating that the surrounding industry is responsible for specific pollution. Because of this, we will have to clarify that these are community results that should not be taken for diagnosis purposes.

## Accessing the Data and Initial Processing

Starting with the PLACES dataset, we can either read it from a csv, downloaded [here](https://chronicdata.cdc.gov/500-Cities-Places/PLACES-Local-Data-for-Better-Health-Place-Data-202/eav7-hnsx) - or read from a url. I have found reading from the url to take too long, so I recommend just downloading the file.

```python
import pandas as pd
from matplotlib import pyplot as plt
from  scipy import stats 
import numpy as np
import warnings
from scipy.stats import f_oneway
warnings.filterwarnings('ignore') # does not like some of my slicing
```

```python
#df_medical = pd.read_csv('https://chronicdata.cdc.gov/api/views/eav7-hnsx/rows.csv?accessType=DOWNLOAD&bom=true&format=true') # Takes a long time to load
df_medical = pd.read_csv('PLACES__Local_Data_for_Better_Health__Place_Data_2020_release.csv')
```

```python
# Then want to make a dataset of the cancer/asthmna data
df_cancerData = df_medical[(df_medical['MeasureId']=='CANCER') & (df_medical['Data_Value_Type']=='Age-adjusted prevalence')]
df_asthmaData = df_medical[(df_medical['MeasureId']=='CASTHMA') & (df_medical['Data_Value_Type']=='Age-adjusted prevalence')]
del df_medical # free up any space

# Select what we care about:
df_cancerData_important = df_cancerData[['StateAbbr','LocationName','Data_Value','Low_Confidence_Limit','High_Confidence_Limit']] 
df_asthmaData_important = df_asthmaData[['StateAbbr','LocationName','Data_Value','Low_Confidence_Limit','High_Confidence_Limit']] 

# To join with the TRI later:
df_cancerData_important['LocationName']= df_cancerData_important['LocationName'].str.lower()
df_asthmaData_important['LocationName']= df_asthmaData_important['LocationName'].str.lower()
```

Can access the TRI from the web, as loading does not take too long:

```python
df_TRI_2018 = pd.read_csv('https://enviro.epa.gov/enviro/efservice/MV_TRI_BASIC_DOWNLOAD/year/=/2018/fname/TRI_2018_US.csv/CSV',low_memory=False)
df_TRI = df_TRI_2018[['6. CITY','8. ST','20. INDUSTRY SECTOR','34. CHEMICAL','42. CARCINOGEN','45. 5.1 - FUGITIVE AIR','46. 5.2 - STACK AIR','47. 5.3 - WATER','54. 5.5.2 - LAND TREATMENT','101. TOTAL RELEASES']]

# To merge with the PLACEs dataset later:
df_TRI['6. CITY'] = df_TRI['6. CITY'].str.lower()
```

## Investigating the Cancer Data

We will start our analysis by looking at the total emissions of carcinogens (regardless of route) for all cities, which we can join to the age-adjusted cancer rate data in the PLACEs dataset. Note that a value of 2*10^-5 lb was added for cities which did not have any carcinogen emissions. This was completed as several semi-log plots will be used. As the logarithm of 0 is undefined, these points would not show up without this adjustment, which would then negate the distribution of thousands of points. As such this small value is used instead, which should be taken as a stand-in for the zero line.

```python
df_carcinogen = df_TRI[df_TRI['42. CARCINOGEN']=='YES']
gb_carcinogen_by_loc = df_carcinogen.groupby(['8. ST','6. CITY'])["101. TOTAL RELEASES"].sum()
df_carcinogen_by_loc = gb_carcinogen_by_loc.reset_index()
df_carcinogen_by_loc['6. CITY'] = df_carcinogen_by_loc['6. CITY'].str.lower()
combined_cancer_loc = pd.merge(df_cancerData_important, df_carcinogen_by_loc, left_on=['StateAbbr','LocationName'],right_on=['8. ST','6. CITY'], how='left')
combined_cancer_loc = combined_cancer_loc.drop(columns=['8. ST','6. CITY'],index=1)
newLabels = ["State","City","Age-adjusted cancer rate (%)", "Lower CI for cancer rate","Upper CI for cancer rate","Total carcinogen release (lb)"]
combined_cancer_loc.columns = newLabels

# If no TRI data exist, then no emissions were reported in the city:
combined_cancer_loc["Total carcinogen release (lb)"] = combined_cancer_loc["Total carcinogen release (lb)"].fillna(0.00002) # filled to 2*10^-5 for visualization purposes
```

#### What do the distributions look like?

We can start our analysis by looking at histograms detailing the distributions, for the cancer rate:

```python
combined_cancer_loc.hist(column = 'Age-adjusted cancer rate (%)',bins=20) # Similar in most places...
plt.xlabel("Age-adjusted cancer rate (%)")
plt.ylabel("Number of cities")
plt.title("Distribution of cancer rates")
plt.xlim(3,8)
plt.show()
```

From this histogram, we can see that the vast majority of cities have cancer rates between 6-7%. 

Whereas for the rate of carcinogen emissions:

```python
x = combined_cancer_loc['Total carcinogen release (lb)'] + 0.00002 # add same offset to make up for 0s not appearing
hist, bins = np.histogram(x, bins=18)
logbins = np.logspace(np.log10(bins[0]),np.log10(bins[-1]),len(bins))
plt.hist(x, bins=logbins)
plt.xscale('log')
plt.yscale('log')
plt.title("Distribution of carcinogen emissions")
plt.xlabel("carcinogen emissions (lb)")
plt.ylabel("Number of cities")
plt.show()
```

Note that this plot is on a log-log scale, such that the number of cities with = no carcinogen emissions dominates (>20,000 cities), which comprises more than the majority of the dataset. While among the cities that have carcinogen emissions the emissions tend to be in the range of,000-100,000 lb.

This indicates that we have a large control group of cities without industrial activity in our dataset, to compare to the cities that do.

### Investigating the Cancer Data - All Emissions

Now that we understand the dataset, we can look at the dependence of the cancer rate on carcinogen emissions via all routes (air, water, land)

```python
combined_cancer_loc.plot.scatter(y='Age-adjusted cancer rate (%)',x='Total carcinogen release (lb)',alpha=.1)
plt.title("Cancer rate over total carcinogen release")
plt.xscale('log')
plt.xlim(0.00001,100000000)
plt.show()
```

Note that the points on the very left side represent cities where no emission data occurs, which as stated before comprises the majority of the dataset.

This plot shows that there does not appear to be a strong correlation between increased carcinogen emission and the cancer rate, in fact the pearson correlation shows:

```python
[r, p] = stats.pearsonr(combined_cancer_loc['Total carcinogen release (lb)'],combined_cancer_loc['Age-adjusted cancer rate (%)'])
print("The Pearson correlation coefficient is: ", r, "and is statistically significant with a p of ", p)
```

That the cancer rate actually decreases with increased emissions, at a 95% confidence level. Of course, this effect size is extremely small, and the fact that it is statistically significant most likely is only the case due to the large amount of data used. However, this does show that there is no clear correlation between emissions and cancer rate.

Another thing that may be interesting, is whether the means of cities with emissions are different than those without. As the age-adjusted cancer rate appears to be more or less normally distributed, we can use an ANOVA:

```python
no_emissions = combined_cancer_loc[combined_cancer_loc['Total carcinogen release (lb)']==0.00002]['Age-adjusted cancer rate (%)']
emissions = combined_cancer_loc[combined_cancer_loc['Total carcinogen release (lb)']!=0.00002]['Age-adjusted cancer rate (%)'] 
f, p = f_oneway(no_emissions,emissions)
print("The p value is",p,"therefore the two population means are not statistically different")
```

This indicates that there is no statistically significant difference between cities that have carcinogen emissions and those that do not, which follows from the trends above.

Note that this is for all carcinogen emissions, grouping together all carcinogens, as well as routes of emissions. It is likely that some carcinogens are more potent than others, while some routes cause more exposure to people. My previous analysis showed that the majority of emissions go to the air. Thinking about how the concentration of a pollutant changes as one moves away from a facility, we know see that the concentration in the air would have a cubed dependence on distance, as the emissions should move in all directions forming a semi-sphere. As such, moving away from the plant quickly dilutes the concentration of any emitted chemical - making them less harmful. Therefore, air emissions most likely only impact people surrounding the facilities and not people on the other side of a town/city. This makes the previous results even less surprising. 

Yet this opens a question for other forms of emission, mainly water emissions, where the main dilution that occurs immediately is as the chemical it enters the body of water, and therefore is not as dependent on how close a person lives to the facility, but the proximity to the body of water

### Investigating the Cancer Data - Emissions to Water

We can then recreate a similar analysis with just the subset of carcinogens emitted to water:

```python
gb_carcinogen_by_loc_water = df_carcinogen.groupby(['8. ST','6. CITY'])['47. 5.3 - WATER'].sum()
df_carcinogen_by_loc_water = gb_carcinogen_by_loc_water.reset_index()
combined_cancer_water = pd.merge(df_cancerData_important, df_carcinogen_by_loc_water, left_on=['StateAbbr','LocationName'],right_on=['8. ST','6. CITY'], how='left')
combined_cancer_water = combined_cancer_water.drop(columns=['8. ST','6. CITY'],index=1)
newLabels = ["State","City","Age-adjusted cancer rate (%)", "Lower CI for cancer rate","Upper CI for cancer rate","Carcinogen releases to water (lb)"]
combined_cancer_water.columns = newLabels

combined_cancer_water["Carcinogen releases to water (lb)"] = combined_cancer_water["Carcinogen releases to water (lb)"].fillna(0.00002)
combined_cancer_water.plot.scatter(y='Age-adjusted cancer rate (%)',x="Carcinogen releases to water (lb)", alpha=.1)
plt.title("Cancer rate over carcinogon releases to water")
plt.xscale('log')
plt.xlim(0.00001,100000000)
plt.show()
```

```python
[r, p] = stats.pearsonr(combined_cancer_water["Carcinogen releases to water (lb)"],combined_cancer_loc['Age-adjusted cancer rate (%)'])
print("The Pearson correlation coefficient is: ", r, "which is not statistically significant with a p combined_cancer_water ", p)
```

This time the correlation is not significant, and visually we can see no clear trend. This seems to indicate that water exposure is not significantly different than air. We can also check to see if the means are the same, assuming a normal distribution again with an ANOVA


```python
no_emissions = combined_cancer_water[combined_cancer_water["Carcinogen releases to water (lb)"]==0.00002]['Age-adjusted cancer rate (%)']
emissions = combined_cancer_water[combined_cancer_water["Carcinogen releases to water (lb)"]!=0.00002]['Age-adjusted cancer rate (%)'] 
f, p = f_oneway(no_emissions,emissions)
print("The p value is",p,"therefore the two population means are not statistically different")
```

Again, the two groups are not different, indicating that reported water emissions are not correlated to cancer rates. This is not super surprising, due to the relatively small magnitude of chemicals emitted this way, and the fact that dilution still occurs via this route.

To see if this is the case for all emissions routes, we can also check land emissions. One would think that these would affect operators/plant workers disproportionately, but not community residents as they would be less likely to be exposed.

## Appendix A: Validating the Total TRI Emission to NOx and PM Emission Correlation 

To check if assuming that the total TRI emissions is a good surrogate for NOx and PM emissions, comparison of the TRI to another dataset was completed.

The [2014 National Emissions Inventory Data](https://edap.epa.gov/public/extensions/nei_report_2014/dashboard.html#point-db) includes this. Note that the version I am using is in the data folder.

Specifically, taking the point summary file and comparing it to the 2014 TRI, we can determine if NOx and PM emissions are correlated with total TRI emissions:

```python
# Reading data from the file
df_NEI_PS = pd.read_csv('Data/NEI_Report_PointSource_Data_2014v2.txt',sep='|')
df_NEI_PS = df_NEI_PS[['STATE','STATE_COUNTY','POLLUTANT','EMISSION_TONS']]
df_NEI_PS['County'] = df_NEI_PS['STATE_COUNTY'].str.extract(r'(\w+$)')
df_NEI_PS['State'] = df_NEI_PS['STATE_COUNTY'].str.extract(r'(^\w+)')
df_NEI_PS['County'] = df_NEI_PS['County'].str.lower()
```

```python
# Want to compare to the 2014 TRI
df_TRI_2014 = pd.read_csv('https://enviro.epa.gov/enviro/efservice/MV_TRI_BASIC_DOWNLOAD/year/=/2014/fname/TRI_2014_US.csv/CSV',low_memory=False)
df_TRI_2014 = df_TRI_2014[['8. ST','7. COUNTY','34. CHEMICAL','101. TOTAL RELEASES']]
df_TRI_2014['7. COUNTY'] = df_TRI_2014['7. COUNTY'].str.lower()
df_2014_all = df_TRI_2014.groupby(['8. ST','7. COUNTY'])['101. TOTAL RELEASES'].sum().reset_index()
```

```python
df_NOx = df_NEI_PS[df_NEI_PS['POLLUTANT']=='Nitrogen Oxides']
df_PM = df_NEI_PS[df_NEI_PS['POLLUTANT']=='PM10']
df_NOx = df_NOx.groupby(['State','County'])['EMISSION_TONS'].sum().reset_index()
df_PM = df_PM.groupby(['State','County'])['EMISSION_TONS'].sum().reset_index()
```

```python
combined_NOx = pd.merge(df_NOx, df_2014_all, left_on=['State','County'],right_on=['8. ST','7. COUNTY'], how='left')
combined_NOx = combined_NOx.dropna() # drops AK due to name conventions 
combined_NOx.plot.scatter(x='101. TOTAL RELEASES',y='EMISSION_TONS',alpha=0.1)
plt.xlabel("Total TRI emissions (lb)")
plt.ylabel("Total NOx point source emissions")
plt.title("NEI NOx emissions over total TRI emissions")
plt.show()
[r, p] = stats.pearsonr(combined_NOx["101. TOTAL RELEASES"],combined_NOx['EMISSION_TONS'])
print("The Pearson correlation coefficient is: ", r, "and is statistically significant with a p of ", p)
```

```python
combined_PM = pd.merge(df_PM, df_2014_all, left_on=['State','County'],right_on=['8. ST','7. COUNTY'], how='left')
combined_PM = combined_PM.dropna()
combined_PM.plot.scatter(x='101. TOTAL RELEASES',y='EMISSION_TONS',alpha=0.1)
plt.xlabel("Total TRI emissions (lb)")
plt.ylabel("Total PM point source emissions")
plt.title("NEI PM emissions over total TRI emissions")
plt.show()
[r, p] = stats.pearsonr(combined_PM["101. TOTAL RELEASES"],combined_PM['EMISSION_TONS'])
print("The Pearson correlation coefficient is: ", r, "and is statistically significant with a p of ", p)
```

The Pearson correlation coefficient of ~0.35 for both of these pollutants shows that there is a positive correlation between these two values, indicating that using the TRI as a surrogate for these two chemicals is not exact, but at least follows the same general trend.
